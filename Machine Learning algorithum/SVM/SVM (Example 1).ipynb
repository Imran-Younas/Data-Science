```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.data
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df["target"] = iris.target

df["flower names"] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df['flower names'].value_counts()
```

```python
df[df["flower names"] == "setosa"  ]
```

```python
df1 = df[0:50]
df2 = df[50:100]
df3 = df[100:150]
```

```python
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color = "green", marker = '+')
plt.scatter(df2["sepal length (cm)"], df2["sepal width (cm)"], color = "blue", marker = "*")

plt.xlabel("sepal length")
plt.ylabel("sepal length")

plt.show()
```

```python
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color = "green", marker = '+')
plt.scatter(df2["petal length (cm)"], df2["petal width (cm)"], color = "blue", marker = "*")

plt.xlabel("petal length")
plt.ylabel("petal length")

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = ['target', 'flower names'])
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X_test
```

```python
model.predict([[1.9,4.1,9.4,1.1]])
```

```python
model_c = SVC(C = 9)   # C use for regulization   
model_c.fit(X_train, y_train)
model_c.score(X_test, y_test)
```

```python
model_g = SVC(gamma=20)   # g use for gamma   
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```
